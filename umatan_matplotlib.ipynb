```python
# for Explanatory Data Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
plt.hist(df["fare"])
plt.show()
```

```python
plt.hist(df["fare"], bins=40)
plt.show()
```

```python
plt.hist(df["fare"], bins=40, range=(0, 100))
plt.show()
```

```python
# Line graph
plt.plot(df['fare'])
plt.show()
```

```python
plt.figure(figsize= (20, 5))
plt.plot(df['fare'])
plt.title('test')
plt.xlabel('data_index')
plt.ylabel('fare')
plt.show()
```

```python
df['fare'].hist(by=df['sex'])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize= (20, 10))
axes[0][0].hist(df['age'], bins= 20)
axes[0][1].hist(df['fare'], bins= 20)
axes[0][1].set_xlim(0, 200)
axes[1][0].hist(df['sex'], bins= 20)
axes[1][1].hist(df['pclass'], bins= 20)
plt.show()
```

```python
df.groupby('pclass').count()
```

```python
# Bar graph (x, y)
plt.bar(df.groupby('pclass').count().index, df.groupby('pclass').count()['survived'])
```

```python
# Scatter plot
plt.scatter(df['fare'], df['age'])
plt.show()
```

```python
plt.scatter(df['fare'], df['age'], alpha=0.2)
plt.show()
```
